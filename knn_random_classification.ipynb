```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
np.random.seed(42)  # for reproducibility
X = np.random.rand(100, 1)
```

```python
y = np.array(['C1' if val <= 0.5 else 'C2' for val in X[:50].flatten()])
```

```python
X_train = X[:50]
X_test = X[50:]
```

```python
k_values = [1, 2, 3, 4, 5, 20, 30]

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y)
    predictions = knn.predict(X_test)
    
    print(f"\n--- Predictions for k = {k} ---")
    for i, val in enumerate(X_test):
        print(f"Test Value {50+i}: {val[0]:.4f}, Predicted Class: {predictions[i]}")
```
